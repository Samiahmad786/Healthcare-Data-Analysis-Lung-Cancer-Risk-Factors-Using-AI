# GroupBy Analysis 
### Smoker vs Non-Smoker Risk

```python
df.groupby('smoker')['lung_cancer_risk'].mean()
```

### Exercise Impact

```python
df.groupby('exercise_hours_per_week')['lung_cancer_risk'].mean()
```

### Passive Smoking Impact

```python
df.groupby('passive_smoking')['lung_cancer_risk'].mean()
```

```python
# STEP 11.1: GroupBy Calculations

# Smoker vs Lung Cancer Risk
smoker_risk = df.groupby('smoker')['lung_cancer_risk'].mean()

# Exercise vs Lung Cancer Risk
exercise_risk = df.groupby('exercise_hours_per_week')['lung_cancer_risk'].mean()

# Passive Smoking vs Lung Cancer Risk
passive_risk = df.groupby('passive_smoking')['lung_cancer_risk'].mean()
```

```python
plt.figure(figsize=(18,5))

# Graph 1: Smoking vs Risk
plt.subplot(1,3,1)
sns.barplot(x=smoker_risk.index, y=smoker_risk.values)
plt.title("Smoking vs Lung Cancer Risk")
plt.xlabel("Smoker (0 = No, 1 = Yes)")
plt.ylabel("Avg Cancer Risk")

# Graph 2: Exercise vs Risk
plt.subplot(1,3,2)
plt.plot(exercise_risk.index, exercise_risk.values, marker='o')
plt.title("Exercise vs Lung Cancer Risk")
plt.xlabel("Exercise Hours / Week")
plt.ylabel("Avg Cancer Risk")

# Graph 3: Passive Smoking vs Risk
plt.subplot(1,3,3)
sns.barplot(x=passive_risk.index, y=passive_risk.values)
plt.title("Passive Smoking vs Lung Cancer Risk")
plt.xlabel("Passive Smoking (0 = No, 1 = Yes)")
plt.ylabel("Avg Cancer Risk")

plt.tight_layout()
plt.show()
```

# Feature Engineering

```python
df['smoking_intensity'] = df['cigarettes_per_day'] * df['smoking_years']
```

```python
df['health_score'] = df['bmi'] + df['crp_level'] - df['exercise_hours_per_week']
```

```python
sns.boxplot(df['crp_level'])
plt.show()
```

# High-Risk Population Segmentation

```python
df.groupby(['age','smoker'])['lung_cancer_risk'].mean().sort_values(ascending=False)
```

```python
pivot_age_smoker = df.pivot_table(
    values='lung_cancer_risk',
    index='age',
    columns='smoker',
    aggfunc='mean'
)

plt.figure(figsize=(6,4))
sns.heatmap(pivot_age_smoker, cmap='Reds')
plt.title("High Risk Population: Age vs Smoker")
plt.xlabel("Smoker (0=No, 1=Yes)")
plt.ylabel("Age")
plt.show()
```

# Smoking vs Non-Smoking BUT Pollution Impact

```python
df.groupby(['smoker','air_pollution_index'])['lung_cancer_risk'].mean()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x='air_pollution_index',
    y='lung_cancer_risk',
    hue='smoker',
    data=df
)
plt.title("Smoking & Pollution Impact on Risk")
plt.show()
```

# Exercise as Risk-Reduction Factor

```python
df.groupby(pd.cut(df['exercise_hours_per_week'], bins=[0,2,5,10]))['lung_cancer_risk'].mean()
```

```python
exercise_bins = pd.cut(df['exercise_hours_per_week'], bins=[0,2,5,10,20])
exercise_trend = df.groupby(exercise_bins)['lung_cancer_risk'].mean()

plt.figure(figsize=(6,4))
exercise_trend.plot(marker='o')
plt.title("Exercise vs Lung Cancer Risk")
plt.xlabel("Exercise Hours (Grouped)")
plt.ylabel("Avg Risk")
plt.show()
```

# Passive Smoking Hidden Risk

```python
df.groupby(['smoker','passive_smoking'])['lung_cancer_risk'].mean()
```

```python
passive_combo = df.groupby(
    ['smoker','passive_smoking']
)['lung_cancer_risk'].mean().reset_index()

plt.figure(figsize=(6,4))
sns.barplot(
    x='smoker',
    y='lung_cancer_risk',
    hue='passive_smoking',
    data=passive_combo
)
plt.title("Passive Smoking Hidden Risk")
plt.xlabel("Smoker (0=No, 1=Yes)")
plt.ylabel("Avg Risk")
plt.show()
```

# Early Warning Health Indicators

```python
df.corr()['lung_cancer_risk'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    x='lung_cancer_risk',
    y='crp_level',
    data=df
)
plt.title("CRP Level vs Lung Cancer Risk")
plt.xlabel("Lung Cancer Risk")
plt.ylabel("CRP Level")
plt.show()
```

# ALL 5 DIAGRAMS IN ONE SINGLE FIGURE

```python
plt.figure(figsize=(20,12))

# 1️⃣ High Risk Segmentation
plt.subplot(3,2,1)
sns.heatmap(pivot_age_smoker, cmap='Reds', cbar=False)
plt.title("1. High Risk Population")

# 2️⃣ Smoking + Pollution Impact
plt.subplot(3,2,2)
sns.scatterplot(
    x='air_pollution_index',
    y='lung_cancer_risk',
    hue='smoker',
    data=df,
    legend=False
)
plt.title("2. Smoking & Pollution Impact")

# 3️⃣ Exercise Trend
plt.subplot(3,2,3)
exercise_trend.plot(marker='o')
plt.title("3. Exercise Risk Reduction")
plt.xlabel("Exercise Hours (Grouped)")
plt.ylabel("Avg Risk")

# 4️⃣ Passive Smoking Risk
plt.subplot(3,2,4)
sns.barplot(
    x='smoker',
    y='lung_cancer_risk',
    hue='passive_smoking',
    data=passive_combo
)
plt.title("4. Passive Smoking Risk")

# 5️⃣ CRP Early Warning
plt.subplot(3,2,5)
sns.boxplot(
    x='lung_cancer_risk',
    y='crp_level',
    data=df
)
plt.title("5. CRP Early Warning")

plt.tight_layout()
plt.show()
```

### Key Insights
- Smoking and pack years strongly increase lung cancer risk
- High air pollution lowers oxygen saturation
- Exercise has a protective effect
- CRP level is a strong inflammation indicator
- Passive smoking also increases cancer risk


### Conclusion
This analysis identifies smoking exposure, air pollution, inflammation, and low physical activity
as major contributors to lung cancer risk. Preventive healthcare policies should focus on
smoking cessation, pollution control, and lifestyle improvement.
